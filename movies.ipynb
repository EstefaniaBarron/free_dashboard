```python
from jupyter_dash import JupyterDash
from dash import dcc, html, Input, Output
import yfinance as yf  
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("rotten_tomatoes_top_movies_2019-01-15.csv")
```

```python
df[df['Genres']=='action|adventure']
```

```python
app = JupyterDash(__name__)
```

```python

app.layout = html.Div(children = [
    html.Div([
        html.H1("Rotten Tomatoes Movie Rankings"),
         
        html.Label("Select Genres:"),
        dcc.Dropdown(
            id='genres-selector',
            options=[
                {'label': 'action|adventure', 'value': 'action|adventure'},
                {'label': 'animation', 'value': 'animation'},
                {'label': 'art_house|international', 'value': 'art_house|international'},
                {'label': 'classics', 'value': 'classics'},
                {'label': 'comedy', 'value': 'comedy'},
                {'label': 'documentary', 'value': 'documentary'},
                {'label': 'drama', 'value': 'drama'},
                {'label': 'horror', 'value': 'horror'},
                {'label': 'kids|family', 'value': 'kids|family'},
                {'label': 'musical|performing_arts', 'value': 'musical|performing_arts'},
                {'label': 'mystery|suspense', 'value': 'mystery|suspense'},
                {'label': 'romance', 'value': 'romance'},
                {'label': 'science_fiction|fantasy', 'value': 'science_fiction|fantasy'},
                {'label': 'special_interest', 'value': 'special_interest'},
                {'label': 'sports|fitness', 'value': 'sports|fitness'},
                {'label': 'television', 'value': 'television'},
                {'label': 'western', 'value': 'western'},
            ],
            value='action|adventure'
        ),
         
        dcc.Graph(id='movie-genre-chart'),
    ]),
    html.Div([
        html.H1("Rotten Tomatoes- Count of Movie Reviews"),
        dcc.Graph(id='movie-reviews-chart'),
    ])
])
        
```

```python
@app.callback(
    Output('movie-genre-chart', 'figure'),
    [Input('genres-selector', 'value')]
)
def update_chart(selected_genre):
    data = df[df['Genres']==selected_genre].head(20)
    fig = px.bar(data, x = "Title", y = "RatingTomatometer", title=f'{selected_genre} Movie Rankings')

    return fig

@app.callback(
    Output('movie-reviews-chart', 'figure'),
    [Input('genres-selector', 'value')]
)
def update_second_chart(selected_genre):
    data = df[df['Genres']==selected_genre].head(20)
    fig = px.bar(data, x = "Title", y = "No. of Reviews", title=f'{selected_genre} Number of Reviews')

    return fig
```
